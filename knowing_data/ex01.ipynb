```python
import pandas as pd
```

```python
data = {
    'Tên': ['John', 'Alice', 'Bob', 'Carol', 'David'],
    'Tuổi': [25, 30, 22, 35, 28],
    'Giới tính': ['Nam', 'Nữ', 'Nam', 'Nữ', 'Nam'],
    'Số điện thoại': ['123-456-7890', '987-654-3210', '555-123-4567', '999-888-7777', '333-444-5555'],
    'Điểm danh gia': [8.5, 9.0, 7.2, 8.9, 6.7]
}
for i in range (len(data)):
    print(data['Tên'][i])
```

```python
df = pd.DataFrame(data)
print(df)
```

```python
# dem so khach hang
print("so luong khach hanh la ",df['Tên'].count())
```

```python
print(df['Giới tính'].count())
```

```python
print('so hoc sinh nam la ',(df['Giới tính'] == 'Nam').sum())
print(df['Giới tính'].value_counts())
```
